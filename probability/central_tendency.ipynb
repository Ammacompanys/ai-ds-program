```python
# Mean
import numpy as np
dataset = [100, 112, 89, 99, 169]
mean = np.mean(dataset)
print(mean)
```

```python
# Median
import numpy as np
dataset = [100, 112, 89, 99, 169]
median = np.median(dataset)
print(median)
```

```python
# Mode
import scipy.stats as sp

dataset = [100, 112, 9, 99, 100]
mode = sp.mode(dataset)
print(mode)
```
