```python
# 1. Range
values = [100, 112, 89, 99, 109]
max_value = max(values)
min_value = min(values)
v_range = max_value - min_value
print(v_range)
```

```python
# 2. Variance
import numpy as np

values = [100, 112, 80, 9, 10]
var = np.var(values)
print("Variance : ", var)
```

```python
# 3. Standard Deviation
import numpy as np

value = [19, 17, 20, 48, 00]
std = np.std(value)
print("Standard Deviation : ", std)
```

```python
# Z-score
import numpy as np
import scipy.stats as sp

data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
zs = sp.zscore(data)
print(zs)
```

```python
# Quartitles
import numpy as np

value = [10, 112, 89, 99, 10]

x = np.quantile(value, [0.25, 0.5, 0.75])
print(x)
```

```python
# InterQuartiles
import numpy as np

value = [10, 112, 89, 99, 10]

Q1 = np.quantile(value, 0.25)
Q3 = np.quantile(value, 0.75)

IQR = Q3 - Q1

print(IQR)
```
