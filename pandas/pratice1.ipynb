```python
import pandas as pd

data = {
	'Department' : ['HR', 'IT', 'HR', 'IT', 'HR', 'Finance'],
	'Employee' : ['A', 'B', 'C', 'E', 'F', 'G'],
	'Salary' : [20000, 10000, 15000, 20000, 60000, 25000]
}

df = pd.DataFrame(data)

grouped = df.groupby("Department")["Salary"].mean()
print(grouped)
result = df.groupby("Department")['Salary'].agg(['mean', 'sum', 'max'])
result2 = df.groupby('Department')['Salary'].mean().reset_index()
print(result2)
```

pivot()

- it converts rows into columns - useful when you want to summarize or display data in a table-like form

```python
import pandas as pd

data = {
	"Month" : ['Jan', 'Jan', 'Feb', 'Feb'],
	"Product" : ['A', 'B', 'A', 'B'],
	"Sales" : [100, 150, 200, 250]
}

df = pd.DataFrame(data)
pivoted = df.pivot(index='Month', columns='Product', values='Sales')

print(pivoted)
```

```python
data = {
	"Rollno" : [1, 2, 3, 6, 8],
	"Sem" : [1, 1, 2, 5, 5],
	"Names" : ["Vijay", "Ajay", "Nithin", "Mithun", "Karan"]
}

df = pd.DataFrame(data)
pivoted_df = df.pivot(index='Rollno', columns='Sem', values='Names')
print(pivoted_df)
```

```python
melted = pivoted_df.reset_index().melt(id_vars='Rollno', var_name="Sem", value_name="Names")
print(melted)
```

- id_vars : variables that identify rows : columns to keep
- var_name : Name for melted variable column : column for original column name
- value_name : Name for melted value column : column for original cell values

```python
import pandas as pd

data = {
	"cus_id" : [1, 2, 4, 5, 6],
	"cus_name" : ["Maco", "Jon", "Mike", "Tyson", "Carl"],
	"item" : ["Milk", "Bread", "Juice", "Cake", "Jam"]
}

df = pd.DataFrame(data)

pivoted = df.pivot(index="cus_id", columns="cus_name", values="item")
melted = pivoted.reset_index().melt(id_vars="cus_id", var_name="cus_name", value_name="item")
print(melted)
```

```python
# filter()

numbers = [31, 4, 6, 71, 8, 11, 17]
result = list(filter(lambda x : x > 20, numbers))
print(result)

df = pd.DataFrame({
	"Name" : ["Suresh", "Mahesh", "Ganesh"],
	"Age" : [20, 21, 22],
	"City" : ["NY", "LA", "SF"]
})

df_filtered = df.filter(["Name", "Age"])
print(df_filtered)

df_filtered = df.filter(like="A")
print(df_filtered)
```

```python
# reduce

from functools import reduce

numbers = [1, 2, 4, 5, 7, 8]
sum_reduce = reduce(lambda a , b : a + b , numbers)
print(sum_reduce)

product_reduce = reduce(lambda a , b : a * b, numbers)
print(product_reduce)

df = pd.DataFrame({
	'A' : [1, 3, 4, 6],
    'B' : [8, 3, 2, 1],
    'C' : [7, 4, 1, 2]
}
)
print(df)

single_col = reduce(lambda x, y: x + y , df["A"])
df["Sum"] = reduce(lambda x , y : x + y, [df["A"], df["B"], df["C"]])
print(df)
```
